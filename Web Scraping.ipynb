**Importing the libraries**

```python
from tqdm.notebook import tqdm as tqdm
from bs4 import BeautifulSoup
import requests
import time
```

**Getting the URL**

```python
ebay_url = 'https://www.ebay.com/sch/i.html?_nkw=lg+phone'
response = requests.get(ebay_url)
```

**Parsing the html page through BeautifulSoup**

```python
soup = BeautifulSoup(response.text, "html.parser")
```

**Saving the HTML Pages**

```python
def saveString(html, filename="text.html"):
	try:
		file = open(filename,"w")
		file.write(str(html))
		file.close()
	except Exception as ex:
		print('Error: ' + str(ex))

saveString(soup, "ebay_lg_phone_01.htm")
```

**Saving All the HTML Pages**

```python
for i in tqdm(range(1,11)):
    url = f'https://www.ebay.com/sch/i.html?_nkw=lg+phone&rt=nc&LH_Auction=1&_pgn={i}'
    response = requests.get(url, headers = header )
    soup = BeautifulSoup(response.text, "html.parser")
    saveString(soup,f'ebay_lg_phone{i}.html')
    time.sleep(10)
```

**Loading the HTML**

```python
def loadString(f="test.html"):
	try:
		html = open(f, "r", encoding='utf-8').read()
		return(html)
	except Exception as ex:
		print('Error: ' + str(ex))
```

**Printing through all the pages and the bids**

```python
for i in tqdm(range(1,11)):
    soup = loadString(f'ebay_lg_phone{i}.html')
    soup = BeautifulSoup(soup, "html.parser")
    bids = soup.findAll('span', class_ = 's-item__bids s-item__bidCount')
    print(f'Page{i} Starts:')
    items = soup.findAll('a', class_ = 's-item__link')
    for j in tqdm(range(0,len(bids))):
        print('The Number of bids:',bids[j].text)
        print('The item link is:',items[j+1]['href'])
```
